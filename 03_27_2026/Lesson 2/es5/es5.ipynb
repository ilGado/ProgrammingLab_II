5. Manipolazione di vettori

5.1: reare un vettore contenente i seguenti stipendi: 50.000, 105.250, 55.000, 89.000.
Qual è il costo totale del personale (somma di tutti gli stipendi dell'azienda)?

```python
arr = np.array([50000, 105250, 55000, 89000])

somma = np.sum(arr)

print(f"Costo totale del personale: {somma}")
```

5.2: Supponiamo che il nostro malvagio CEO abbia deciso di darsi un aumento.
Modifica il vettore degli stipendi in modo che il CEO, la persona che guadagna 105.250 dollari, riceva un aumento del 15%.

```python
arr[1] += arr[1]/100 * 15

print(arr) 
```

5.4 Ricrea il vettore e utilizza l'argomento dtype in modo che, quando il CEO riceve un aumento del 15%, il suo stipendio finale sia esattamente 121.037,50 dollari.

```python
newArr = np.array([50000, 105250, 55000, 89000], dtype=np.float64) 
print(newArr)

newArr[1] += newArr[1]/100 * 15

print(newArr) 
```

5.5 Questo aumento ha infastidito l'impiegata con lo stipendio più basso (50.000 dollari), che ora pretende un aumento del 20%.
Modifica il vettore per riflettere questo cambiamento.

```python
newArr[0] += newArr[0]/5

print(newArr) 
```

5.6 L'aumento dato al CEO e all'impiegata ha irritato gli altri due dipendenti, quindi ora devono ricevere un aumento del 10%.
Modifica il vettore di conseguenza.

```python
newArr[2] += newArr[2]/10

newArr[3] += newArr[3]/10

print(newArr)
```

5.7 Calcola il costo totale del personale dopo tutti gli aumenti.
Alla fine, quanto è costato all'azienda l'aumento di circa 16.000 dollari del CEO?

```python
newSum = sum(newArr)

print(f"nuovo costo del personale: {newSum}")

print(f"Incremento del: {newSum - somma}")
```
