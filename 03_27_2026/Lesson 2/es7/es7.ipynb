7. Analisi di una matrice

Data la matrice survey con età, reddito e anni di istruzione


```python
survey_matrix = np.array([ [25, 40000, 10], [32, 52000, 12], [40, 63000, 14], [29, 47000, 11], [35, 58000, 13] ])

print(survey_matrix)
```

7.1 Seleziona dalla matrice solo gli intervistati con almeno 12 anni di istruzione utilizzando un test logico.

```python
print(survey_matrix[survey_matrix[:,2] >= 12])
```

7.2 In un'unica riga di codice, seleziona i redditi degli intervistati con almeno 12 anni di istruzione.



```python
print(survey_matrix[survey_matrix[:,2] >= 12][:,1])
```

7.3 In un'unica riga di codice, calcola il reddito medio degli intervistati con almeno 12 anni di istruzione.

```python
print(sum(survey_matrix[survey_matrix[:,2] >= 12][:,1])/len(survey_matrix[survey_matrix[:,2] >= 12][:,1]))
```
