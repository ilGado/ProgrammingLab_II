1. Indicizzazione booleana:

1.1: Creare un nuovo vettore con tutti i numeri primi tra 0 e 20. I numeri primi compresi tra 0 e 20 sono: 2, 3, 5, 7, 11, 13, 17, 19.


```python
def is_prime(x):
    
    for i in range(2, x - 1):
        
        if x % i == 0:
            return False
        
    return True


arr = np.array([i for i in range (2, 20) if is_prime(i)])
print(f"Array: {arr}")



```

1.2: Utilizza un test logico (ad esempio >, <, ==, !=) per selezionare tutte le voci del vettore maggiori di 10.


```python
print(f"Numeri > 10: {arr[arr > 10]}")
```

1.3: Utilizza un test logico per selezionare tutti i numeri primi pari dalla lista.

```python
print(f"Numeri pari: {arr[arr % 2 == 0]}")
```
