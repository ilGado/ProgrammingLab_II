1.4: Scrivi l'array usando una list comprehension che controlla che i numeri siano primi.

```python
def is_prime(x):
    for i in range(2, x - 1):
        if x % i == 0:
            return False
        
    return True

print(is_prime(10))
primes = np.array([i for i in range(2, 10) if is_prime(i)])
print(primes)
```
